# Summary

The temperature structure of the lithosphere affects a wide variety of physical properties and processes. For instance, rock density and rheology are temperature-dependent, which impacts lithospheric deformation and evolution over geological times.

|   |
|---|
| <img src="images/earth_layering.png" alt="Lithosphere" width="400"/> |
| ***Figure 1:** The outer shells of the Earth ([Stüwe 2002](https://doi.org/10.1007/978-3-540-71237-4))* |

There are three mechanisms for heat transfer: conduction, convection, and radiation. Conduction and convection are the most dominant mechanisms when studying solid Earth processes over geological times. The temperature distribution in the continental lithosphere is governed mainly by the conductive heat loss to the surface of heat that is generated internally in the crust by the decay of radioactive isotopes (e.g., U, Th, and K) and the upward heat flow from the sublithospheric mantle. In the oceanic lithosphere, heat loss is also largely controlled by conduction, although the contribution of convective heat transport by water circulating through the basaltic crustal rocks should be considered, especially near mid-oceanic ridges. Heat transfer below the lithosphere is through convective heat transfer through which mantle convection is possible. 

## Conductive heat transfer equations
The governing differential equation of time-dependent conductive heat transfer is: 

$$-div (-k\cdot grad(T)) + A =  \rho c \frac{\delta T}{\delta t} \tag{1}$$

where $T$ $(K)$ is temperature, $t$ $(s)$ is time, $A$ $(W\cdot kg^{-1})$ is volumetric heat production, $k$ $(W\cdot m^{-1}\cdot K^{-1})$ is thermal conductivity, $\rho$ $(kg\cdot m^{-3})$ is density and $c$ $(J \cdot kg^{-1} \cdot K^{-1})$ is specific heat.

For one-dimensional heat transfer in a  homogeneous and isotropic medium, equation (1) can be written as:

$$ \frac{\delta^{2} T}{\delta z^{2}} + \frac{A}{K} = \frac{1}{\alpha} \frac{\delta T}{\delta t} \tag{2}$$

where $\alpha = \frac{k}{\rho c}$ is the thermal diffusivity ($m^{2} \cdot s^{-1}$).

Simplification of equation (2) assuming a steady state heat transfer leads to Poisson's equation:

$$ \frac{\delta^{2}T}{\delta z^{2}} = - \frac{A}{k} \tag{3}$$

The solution of equation (3) in a layer of constant heat generation and constant thermal conductivity is:

$$ T(z) = T_{0} + \frac{q_{0}}{k} z - \frac{Az^{2}}{2k} \tag{4}$$

where $T_{0}$ and $q_{0}$ ($W\cdot m^{-2}$) are the temperature and heat flow respectively at the top of the layer. 

If the layer has thickness z, then the temperature at, and heat flow through, the bottom of the layer $(T_{b}, q_{b})$ can be expressed in terms of the temperature and heat flow at the top of the layer $(T_{0}, q_{0})$ and properties (A, k) of the layer:

$$ T_{b} = T_{0} + \frac{q_{0}}{k} z - \frac{A\Delta z^{2}}{2k} \tag{5}$$

$$ q_{b} = q_{0} - A\Delta z \tag{6}$$

Equations (5) and (6) can be applied to successive layers, resetting $T_{0}$ and  $q_{0}$ at the top of each new layer with the values $T_{b}$ and  $q_{b}$ solved for the bottom of the previous layer.

## Thermal parameters

Equations (5) and (6) indicate that the Earth’s geothermal gradient is controlled by the thermal conductivity (k) and the volumetric heat production (A) of rocks.

**Thermal conductivity (k)** of most rocks at crustal conditions varies inversely with temperature and directly with pressure (or depth):

$$ k(T, z) = \frac{k_{0}(1+cz)}{1+bT} \tag{7}$$

where T ($\degree C$) is temperature, $b$ and $c$ are constants, and $k_{0}$ is conductivity measured at 0°C and 1 atm (101325 Pa).

Given the inherent heterogeneity of the Earth’s crust and, to a lesser degree, mantle, it is necessary to simplify the compositional layering of the lithosphere, when computing the geotherm at crustal or lithospheric scale. Often, generalized lithosphere models assume homogeneous layering with felsic (e.g., granite) and/or mafic (e.g., gabbro, granulite) composition in the crust, and olivine composition in the mantle.

Volumetric heat production (A) in the crust and mantle is due to the radioactive decay of $^{235}U$, $^{238}U$, $^{232}Th$, and $^{40}K$ and constitute a substantial part of the heat lost through the Earth’s surface.

The total heat production $(A_{0})$ is related to the heat generation rates of the individual radioactive elements:

$$ A = C^{U} A^{U} + C^{Th} A^{Th} + C^{K} A^{K} \tag{8}$$

where C is the concentration of the elements and A their respective rate of heat release.

The concentration (C) of a radioactive isotope at time (t) measured backward from the present is related to present day concentration $(C_{0})$ and the half-life of the isotope $(\tau_{1/2})$:

$$ C = C_{0} \exp{\left ( \frac{t\ln (2)}{\tau_{1/2}} \right )} \tag{9}$$

Unlike the mantle (Table 2), the concentrations of heat-producing elements in the crust vary considerably (Table 3). As for thermal conductivity, it is again necessary to use a simplified model for the calculation of geotherms. This should take into account the following: 
- average levels of heat generation measured on surface felsic rocks cannot be sustained throughout the crust because the heat flow produced by such a vertical column would exceed the surface heat flow of $40 mW\cdot m^{-2}$ characteristic of shields; 
- a vertical distribution of heat generation in the crust should satisfy the linear heat flow-heat production relationship.

For the upper crust an exponentially decreasing heat production is used $A(z) = A_{0} \exp{\left (\frac{-z}{D} \right )}$, which satisfies the linear heat flow-heat production relationship:

$$ q_{0} = q_{r} + DA_{0} \tag{10}$$

where $q_{0}$ is surface heat flow, $q_{r}$ is reduced heat flow, and $A_{0}$ is surface heat generation. D is a parameter with dimensions of length and characterizes the depth distribution A(z).

In order to parameterize the geotherm calculation in terms of surface heat flow, an empirical relationship that partitions the observed surface heat flow, with 40% being attributed to upper crustal radiogenic sources and 60% to deeper sources, is often used ([Pollack and Chapman 1977](https://doi.org/10.1016/0040-1951(77)90215-3); [Vitorello and Pollack 1980](https://doi.org/10.1029/JB085iB02p00983); [Chapman 1986](https://doi.org/10.1144/GSL.SP.1986.024.01.07)). Thus surface radioactivity $A_{0}$ can be related simply to surface heat flow using $A_{0} = 0.4 \frac{q_{0}}{D}$. The exponential distribution is continued downwards until the heat production falls to the lower crust heat production $A_{L}$ at $z = D + \ln{\left (\frac{A_{0}}{A_{L}} \right )}$.

The geothermal gradient data is saved as a .csv file under files (files/geotherms.csv). In the code cell below we will:
1. import the data into a DataFrame,
2. interpolate the data, and

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

# Read the CSV file into a DataFrame
df = pd.read_csv('files/geotherms.csv')
display(df)
````

````python
# Create a new DataFrame to store the interpolated data
df_interpolated = pd.DataFrame(columns=df.columns)

# Interpolate the data every 1 km
df_interpolated['z(km)'] = range(0, 41, 1)

df_interpolated['P(MPa)'] = np.interp(df_interpolated['z(km)'], df['z(km)'], df['P(MPa)'])
df_interpolated['T_A(C)'] = np.interp(df_interpolated['z(km)'], df['z(km)'], df['T_A(C)'])
df_interpolated['T_B(C)'] = np.interp(df_interpolated['z(km)'], df['z(km)'], df['T_B(C)'])

display(df_interpolated)
````

````python
# Plot interpolated geothermal gradients 
fig, ax = plt.subplots()
ax.plot(df_interpolated['T_A(C)'], df_interpolated['z(km)'], label=''r'$\nabla T_{A}$')
ax.plot(df_interpolated['T_B(C)'], df_interpolated['z(km)'], label=''r'$\nabla T_{B}$')
ax.set_xlim(0, df_interpolated[['T_A(C)', 'T_B(C)']].max().max()) # Set x-axis limits
ax.set_ylim(df_interpolated['z(km)'].max(), 0)  # Set y-axis limits
ax.set_xlabel('Temperature (C)')
ax.set_ylabel('Depth (km)')
ax.legend()

# Create a function to convert depth to pressure
def depth_to_pressure(depth):
    return np.interp(depth, df_interpolated['z(km)'], df_interpolated['P(MPa)'])

# Create a function to convert pressure to depth
def pressure_to_depth(pressure):
    return np.interp(pressure, df_interpolated['P(MPa)'], df_interpolated['z(km)'])

# Create a secondary y-axis to show pressure
secax_y = ax.secondary_yaxis('right', functions=(depth_to_pressure, pressure_to_depth))
secax_y.set_ylabel('Pressure (MPa)')

plt.show()
````

Using equation (7):
$$ k(T, z) = k_{0}\frac{1+cz}{1+bT} \tag{7}$$

let's investigate how the thermal conductivity of two rock types, granite and granulite, varies with depth assuming the two geothermal gradients above ($\nabla T_{A}$ and $\nabla T_{B}$) and the parameter values in Table 5 below:

***Table 5:** Themal conductivity parameter values ([Chapman 1986](https://doi.org/10.1144/GSL.SP.1986.024.01.07)).*


|     | Granite | Granulite |
|:----|:-------:|:---------:|
| $K_{0}$ $(W\cdot m^{-1} \cdot K^{-1})$ | 3.0 | 2.6 | 
| b $(K^{-1})$ | $1.5 \times 10^{-3}$ | $1.0 \times 10^{-3}$ |
| c $(km^{-1})$ | 1.5 | 1.5 |

````python
# Thermal conductivity parameters
k0_granite = 3  # Granite thermal conductivity at 0C
k0_granulite = 2.6  # Granulite thermal conductivity at 0C
b_granite = 1.5e-3  # Granite thermal conductivity temperature coefficient
b_granulite = 1.0e-3  # Granulite thermal conductivity temperature coefficient
c_granite = 1.5e-3  # Granite thermal conductivity pressure coefficient
c_granulite = 1.5e-3  # Granulite thermal conductivity pressure coefficient
````

````python
## METHOD 1: Calculate thermal conductivities using equation (7) directly

# Copy the previous DataFrame to a new one
df_conductivity_1 = df_interpolated.copy()

# Add four new columns to the new DataFrame to store the calculated conductivities
df_conductivity_1['k_A_granite(W/mK)'] = np.nan  # Granite thermal conductivity along geotherm A
df_conductivity_1['k_B_granite(W/mK)'] = np.nan  # Granite thermal conductivity along geotherm B
df_conductivity_1['k_A_granulite(W/mK)'] = np.nan  # Granulite thermal conductivity along geotherm A
df_conductivity_1['k_B_granulite(W/mK)'] = np.nan  # Granulite thermal conductivity along geotherm B 

# Calculate the thermal conductivities
df_conductivity_1['k_A_granite(W/mK)'] = k0_granite * (1 + c_granite * df_conductivity_1['z(km)']) / (1 + b_granite * (df_conductivity_1['T_A(C)'])) # calculate granite thermal conductivity along geotherm A
df_conductivity_1['k_B_granite(W/mK)'] = k0_granite * (1 + c_granite * df_conductivity_1['z(km)']) / (1 + b_granite * df_conductivity_1['T_B(C)']) # calculate granite thermal conductivity along geotherm B
df_conductivity_1['k_A_granulite(W/mK)'] = k0_granulite * (1 + c_granulite * df_conductivity_1['z(km)']) / (1 + b_granulite * df_conductivity_1['T_A(C)']) # calculate granulite thermal conductivity along geotherm A
df_conductivity_1['k_B_granulite(W/mK)'] = k0_granulite * (1 + c_granulite * df_conductivity_1['z(km)']) / (1 + b_granulite * df_conductivity_1['T_B(C)']) # calculate granulite thermal conductivity along geotherm B

display(df_conductivity_1)

# Plot the thermal conductivities
fig, ax = plt.subplots()
ax.plot(df_conductivity_1['k_A_granite(W/mK)'], df_conductivity_1['z(km)'], color='red', linestyle='-', label=''r'$k_{A, granite}$')
ax.plot(df_conductivity_1['k_B_granite(W/mK)'], df_conductivity_1['z(km)'], color='red', linestyle='--', label=''r'$k_{B, granite}$')
ax.plot(df_conductivity_1['k_A_granulite(W/mK)'], df_conductivity_1['z(km)'], color='green', linestyle='-', label=''r'$k_{A, granulite}$')
ax.plot(df_conductivity_1['k_B_granulite(W/mK)'], df_conductivity_1['z(km)'], color='green', linestyle='--', label=''r'$k_{B, granulite}$')
ax.set_xlim(df_conductivity_1[['k_A_granite(W/mK)', 'k_B_granite(W/mK)', 'k_A_granulite(W/mK)', 'k_B_granulite(W/mK)']].min().min(), df_conductivity_1[['k_A_granite(W/mK)', 'k_B_granite(W/mK)', 'k_A_granulite(W/mK)', 'k_B_granulite(W/mK)']].max().max()) # Set x-axis limits
ax.set_ylim(df_conductivity_1['z(km)'].max(), 0)  # Set y-axis limits
ax.set_xlabel('Thermal Conductivity (W/mK)')
ax.set_ylabel('Depth (km)')
ax.legend()
ax.set_title('Thermal Conductivity vs Depth')

# Create a secondary y-axis to show pressure
secax_y = ax.secondary_yaxis('right', functions=(depth_to_pressure, pressure_to_depth))
secax_y.set_ylabel('Pressure (MPa)')

plt.show()
````

````python
## METHOD 2: calculate thermal conductivities using a function based on equation (7)

# Copy the previous DataFrame to a new one
df_conductivity_2 = df_interpolated.copy()

# Create a function to calculate the thermal conductivity
def calculate_thermal_conductivity(T, z, k0, b, c):
    return k0 * (1 + c * z) / (1 + b * T)

# Calculate the thermal conductivities using the function
df_conductivity_2['k_A_granite(W/mK)'] = calculate_thermal_conductivity(df_conductivity_2['T_A(C)'], df_conductivity_2['z(km)'], k0_granite, b_granite, c_granite)
df_conductivity_2['k_B_granite(W/mK)'] = calculate_thermal_conductivity(df_conductivity_2['T_B(C)'], df_conductivity_2['z(km)'], k0_granite, b_granite, c_granite)
df_conductivity_2['k_A_granulite(W/mK)'] = calculate_thermal_conductivity(df_conductivity_2['T_A(C)'], df_conductivity_2['z(km)'], k0_granulite, b_granulite, c_granulite)
df_conductivity_2['k_B_granulite(W/mK)'] = calculate_thermal_conductivity(df_conductivity_2['T_B(C)'], df_conductivity_2['z(km)'], k0_granulite, b_granulite, c_granulite)

display(df_conductivity_2)

# Plot the thermal conductivities
fig, ax = plt.subplots()
ax.plot(df_conductivity_2['k_A_granite(W/mK)'], df_conductivity_2['z(km)'], color='red', linestyle='-', label=''r'$k_{A, granite}$')
ax.plot(df_conductivity_2['k_B_granite(W/mK)'], df_conductivity_2['z(km)'], color='red', linestyle='--', label=''r'$k_{B, granite}$')
ax.plot(df_conductivity_2['k_A_granulite(W/mK)'], df_conductivity_2['z(km)'], color='green', linestyle='-', label=''r'$k_{A, granulite}$')
ax.plot(df_conductivity_2['k_B_granulite(W/mK)'], df_conductivity_2['z(km)'], color='green', linestyle='--', label=''r'$k_{B, granulite}$')
ax.set_xlim(df_conductivity_2[['k_A_granite(W/mK)', 'k_B_granite(W/mK)', 'k_A_granulite(W/mK)', 'k_B_granulite(W/mK)']].min().min(), df_conductivity_2[['k_A_granite(W/mK)', 'k_B_granite(W/mK)', 'k_A_granulite(W/mK)', 'k_B_granulite(W/mK)']].max().max()) # Set x-axis limits
ax.set_ylim(df_conductivity_2['z(km)'].max(), 0)  # Set y-axis limits
ax.set_xlabel('Thermal Conductivity (W/mK)')
ax.set_ylabel('Depth (km)')
ax.legend()
ax.set_title('Thermal Conductivity vs Depth')



# Create a secondary y-axis to show pressure
secax_y = ax.secondary_yaxis('right', functions=(depth_to_pressure, pressure_to_depth))
secax_y.set_ylabel('Pressure (MPa)')

plt.show()

````

## Mantle heat production

Most of the heat lost through the Earth's surface originates from the interior of the planet by the decay of radioactive elements.

The heat generation per unit mass is given by:

$$ A = \frac{Q}{M} \tag{11} $$

where Q (W) is the heat flow and M (kg) is the mass.

if we assume the following parameters ([Turcotte and Schubert 2014](https://doi.org/10.1017/CBO9780511843877)):
- total continental area: $ S_{c} = 2 \times 10^{8} km^{2} $
- mean continental surface heat flow: $ q_{c} = 65 mW \cdot m^{2} $
- total oceanic area: $ S_{o} = 3.1 \times 10^{8} km^{2} $
- mean oceanic surface heat flow: $ q_{o} = 101 mW \cdot m^{2} $

We can calculate the present total heat lost through the Earth's surface:

````python
# Parameters:
area_c = 2.e8  # Continental area in km^2
q_c = 0.065  # Mean continental surface heat flow in W/m^2
area_o = 3.1e8  # Oceanic area in km^2
q_o = 0.101  # Mean oceanic surface heat flow in W/m^2

# Total Earth heat loss:
Qc = 2.e8 * 1.e6 * 0.065  # Continental heat loss in W
Qo = 3.1e8 * 1.e6 * 0.101  # Oceanic heat loss in W
Q = Qc + Qo  # Total Earth heat loss in W

print('The continental heat loss is {:.2e} W \nThe oceanic heat loss is {:.2e} W \nThe total Earth heat loss is {:.2e} W'.format(Qc, Qo, Q))
````

We can calculate the present day heat generation in the Earth (A), assuming the mass of the Earth: $ M_{Earth} = 5.97 \times 10^{24} kg $

````python
M_earth = 5.97e24  # Earth mass in kg

A = Q / M_earth # Heat flow per unit mass in W/kg

print('The Earth\'s total heat generation is {:.2e} W/kg'.format(A))
````

We know, based on geochemical studies, that the Earth's core do not contain a significant amount of heat-producing elements. Thus, we can subtract the core mass:

````python
M = 4.e24  # Mass of the mantle in kg

A = Q / M  # Heat flow per unit mass in W/kg

print('The Earth\'s total heat generation is {:.2e} W/kg'.format(A))
````

The continental heat flux value used above ($ q_{c} = 65 mW \cdot m^{2} $) includes the heat produced in the mantle and conducted through the crust and the heat generated by heat-producing elements in the crust. Based on estimates of mean concentrations of these elements, we can attribute $ 37 mW\cdot m^{2} $ to heat producing elements and $ 28 mW\cdot m^{2} $ to basal heating by mantle convection.

We can adjust the total heat generation calculated above accordingly to obtain the mantle heat production: 

````python
# Parameters:
area_c = 2.e8  # Continental area in km^2
q_cm = 0.028  # Mean continental basal heat flow in W/m^2
area_o = 3.1e8  # Oceanic area in km^2
q_o = 0.101  # Mean oceanic surface heat flow in W/m^2

M = 4.e24  # Mass of the mantle in kg

# Total Earth heat loss:
Qcm = 2.e8 * 1.e6 * 0.028  # Continental mantle heat loss in W
Qo = 3.1e8 * 1.e6 * 0.101  # Oceanic heat loss in W
Qm = Qcm + Qo  # Total Earth mantle heat loss in W

# Mantle total heat generation

Am = Qm / M  # Heat flow per unit mass in W/kg

print('The Earth\'s mantle total heat generation is {:.2e} W/kg'.format(Am))
````

One additional correction to the above calculated radiogenic heat production in the mantle is required because of the contribution of the secular cooling of the Earth to the surface heat flow. It is assumed that 20% of the present day surface heat flow comes from the cooling of the Earth and 80% can be attributed to radioactively generated heat:

````python
Am = Am * 80 / 100

print('The present-day mantle heat generation is {:.2e} W/kg'.format(Am))
````

Radioactive heat generated in the mantle and crust is due to the decay of U ($^{235}U$ and $^{232}U$), Th ($^{232}Th$), and P ($^{40}K$) isotopes.

Assuming equations (8) and (9), and isotope parameter values from Table 2, we can calculate the mean mantle heat production due to the decay of U, Th, and K function of time measured back from the present:

$$ A = C^{U} A^{U} + C^{Th} A^{Th} + C^{K} A^{K} \tag{8}$$

$$ C = C_{0} \exp{\left ( \frac{t\ln (2)}{\tau_{1/2}} \right )} \tag{9}$$


**(Note that the cell below will take few minutes to run)**

````python
# Parameters:
## Uranium-238
A0_238U = 9.46e-5  # Uranium-238 heat generation in W/kg
tau_238U = 4.47e9  # Uranium-238 decay constant in years
C_238U = 30.8e-9  # Uranium-238 concentration in the mantle in kg/kg
## Uranium-235
A0_235U = 5.69e-4  # Uranium-235 heat generation in W/kg
tau_235U = 7.04e8  # Uranium-235 decay constant in years
C_235U = 0.22e-9  # Uranium-235 concentration in the mantle in kg/kg
## Thorium-232
A0_232Th = 2.64e-5  # Thorium-232 heat generation in W/kg
tau_232Th = 1.4e10  # Thorium-232 decay constant in years
C_232Th = 124e-9  # Thorium-232 concentration in the mantle in kg/kg
## Potassium-40
A0_40K = 2.92e-5  # Potassium-40 heat generation in W/kg
tau_40K = 1.25e9  # Potassium-40 decay constant in years
C_40K = 36.9e-9  # Potassium-40 concentration in the mantle in kg/kg

# Calculate the heat generation through time (4.5 Ga)
# Time range in years
t = np.linspace(0, 4.5e9, 100000000)

def A_isotope(t, A0, C, tau):
    return A0 * C * np.exp(t * np.log(2) / tau)

# Calculate heat generation for Uranium-238
A_U238 = A_isotope(t, A0_238U, C_238U, tau_238U)

# Calculate heat generation for Uranium-235
A_U235 = A_isotope(t, A0_235U, C_235U, tau_235U)

# Calculate heat generation for Thorium-232
A_Th232 = A_isotope(t, A0_232Th, C_232Th, tau_232Th)

# Calculate heat generation for Potassium-40
A_K40 = A_isotope(t, A0_40K, C_40K, tau_40K)

# Plot the heat generation for all isotopes
fig, ax = plt.subplots()
ax.plot(t / 1e9, A_U238, label='238U')
ax.plot(t / 1e9, A_U235, label='235U')
ax.plot(t / 1e9, A_Th232, label='232Th')
ax.plot(t / 1e9, A_K40, label='40K')
ax.plot(t / 1e9, A_U238 + A_U235 + A_Th232 + A_K40, label='Total')
ax.set_xlim(4.5, 0)
ax.set_xlabel('Time (Ga)')
ax.set_ylabel('Heat Generation (W/kg)')
ax.legend()
ax.set_title('Mantle heat Generation vs Time')

plt.show()
````

### Geothermal gradient calculation

It is evident form the data shown above that surface heat flow show a large variability. This highlights the heterogenous nature of the Earth's subsurface, particularly in continental regions.

Using surface heat flow in equations (5) and (6), and making assumptions about some of the thermal parameters, one can calculate the geothermal gradient in the lithosphere:

$$ T_{b} = T_{0} + \frac{q_{0}}{k} z - \frac{A\Delta z^{2}}{2k} \tag{5}$$

$$ q_{b} = q_{0} - A\Delta z \tag{6}$$

|   |
|---|
| <img src="images/geotherm_lithosphere_structure.png" alt="Lithosphere" width="400"/> |
| ***Figure 2:** The assumed lithospheric structure used for calculating the geothermal gradient* |


**Known variables**: 
- T[0]: surface temperature (273 $K$);
- T[3]: temperature at the base of the lithosphere (1613 $K$); 
- q[0]: surface heat flow ($Wm^{-2}$);
- q[3]: heat flow at the base of the lithosphere ($Wm^{-2}$); 

- $\Delta z$: thicknesses of lithospheric layers;
- A[2], A[3]: radiogenic heat productions for lower crust and mantle lithosphere (negligible)  ($Wm^{-3}$).
- k: thermal conductivity assumed constant (see Table 1; $Wm^{-1}K^{-1}$).

**Unknown variables**: 
- T[1]; T[2]: temperatures at the top/base of the lower crust;
- A[1]: radiogenic heat production in the upper crust;
- q[1], q[2]: heat flows at the top of lithospheric layers.

````python
# Define geotherm variables. 
# Layers are [lithosphere,upper crust,middle crust,lower crust,mantle lithosphere]
k   = 2.5                                         # Thermal conductivity W/(m*K)
dz  = np.array([120.e3,20.e3,20.e3,80.e3])         # Layer thickness (m): lithosphere, upper crust, lower crust, mantle lithosphere
A   = np.array([float('nan'),1.e-6,0.25e-6,0.])   # Radiogenic heat production (W/m**3): dummy, upper crust, lower crust, mantle lithosphere
T   = np.array([273.,0.,0.,1613])                 # Temperature at top of layers (K)
q   = np.array([0.1,0.,0.,0.])                    # Heat flow at top of layers (W/m**2)
ATG = 0                                           # Asthenospheric thermal gradient (k/m) - e.g., 0.25e-3

# depth values
z_max = dz[0]    # maximum depth
zi = 0.5e3       # depth increment

z  = np.array(range(0,int(z_max/zi)+1,1))*0.5e3     # depth from 0 to z_max (0.5 km increments)
t  = np.array(range(0,int(z_max/zi)+1,1))*0.        # array for temperature values

# Calculate radiogenic heat production in the upper crust A[1]
A [1] = 2*k/(dz[1]*(2*dz[0]-dz[1]))*(T[0] - T[3] + (q[0]*dz[0] -A[2]*dz[2]*dz[3])/k - (A[2]*dz[2]**2 + A[3]*dz[3]**2)/(2*k))

# Calaculate surface heat flows at boundaries between layers
q[1] = q[0] - A[1]*dz[1]
q[2] = q[1] - A[2]*dz[2]
q[3] = q[2] - A[3]*dz[3]


# Calculate temperatures at top of layers
T[1] = T[0] + (q[0]/k)*dz[1] - (A[1]*dz[1]**2)/(2.*k)
T[2] = T[1] + (q[1]/k)*dz[2] - (A[2]*dz[2]**2)/(2.*k)
T[3] = T[2] + (q[2]/k)*dz[3] - (A[3]*dz[3]**2)/(2.*k)

# Print thermal values
# create a dataframe to print the results
import pandas as pd

# create a dictionary of variables and values to be displayed
thermal_values = {'Layer': ['Upper crust', 'Lower crust', 'Lithospheric mantle', 'Lithosphere'],
                    'Thickness (km)': [dz[1]/1.e3, dz[2]/1.e3, dz[3]/1.e3, dz[0]/1.e3],
                    'Heat production (W/m3)': [A[1], A[2], A[3], float('nan')],
                    'Heat flow (mW/m2)': [q[0]*1.e3, q[1]*1.e3, q[2]*1.e3, float('nan')],
                    'Top temperature (K)': [T[0], T[1], T[2], float('nan')],
                    'Base temperature (K)': [T[1], T[2], T[3], float('nan')]}

# create a dataframe
thermal_values_df = pd.DataFrame(thermal_values)

# print the dataframe
display(thermal_values_df)

# Calculate temperature as a function of depth
for i in range(np.size(z)):
  if z[i]<=dz[1]:
    t[i] = T[0] + (q[0]/k)*z[i] - (A[1]*(z[i]**2))/(2*k)
  elif z[i]>dz[1] and z[i]<=(dz[1]+dz[2]):
    t[i] = T[1] + (q[1]/k)*(z[i]-dz[1]) - (A[2]*((z[i]-dz[1])**2))/(2*k)
  elif z[i]>(dz[1]+dz[2]) and z[i]<=(dz[1]+dz[2]+dz[3]):
    t[i] = T[2] + (q[2]/k)*(z[i]-dz[1]-dz[2]) - (A[3]*((z[i]-dz[1]-dz[2])**2))/(2*k)
  elif z[i]>(dz[1]+dz[2]+dz[3]):
    t[i] = T[3] + ATG*(z[i]-dz[1]-dz[2]-dz[3])

# Plot results
plt.figure()
plt.plot(t,z/1.e3,'r')
plt.xlabel('Temperature (K)')
plt.ylabel('Depth (km)')
plt.title('Geothermal gradient')
plt.ylim(z_max/1.e3,min(z)/1.e3)
plt.xlim(min(t),max(t))
plt.grid(True)
plt.show()
````
