## Heat generated by dominant radioactive elements in mantle

Radioactive heat generated in the mantle and crust is due to the decay of U ($^{235}U$ and $^{232}U$), Th ($^{232}Th$), and P ($^{40}K$) isotopes.

Assuming equations (7) and (8), and isotope parameter values from Table 2, we can calculate the mean mantle heat production due to the decay of U, Th, and K function of time measured back from the present:

$$ A = C^{U} A^{U} + C^{Th} A^{Th} + C^{K} A^{K} \tag{8}$$

$$ C = C_{0} \exp{\left ( \frac{t\ln (2)}{\tau_{1/2}} \right )} \tag{9}$$

**Table 2:** Typical concentrations of the heat-producing elements in several rock types (Table 4.3;  [Turcotte and Schubert 2014](https://doi.org/10.1017/CBO9780511843877)).

| Rock type | U (ppm) | Th (ppm) | K (%) |
|:----------|:-------:|:--------:|:-----:|
| Reference underpleted (fertile) mantle | 0.031 | 0.124 | 0.031 |
| Depleted peridotite | 0.001 | 0.004 | 0.003 |
| Tholeiitic basalt | 0.07 | 0.19 | 0.088 |
| Granite | 4.7 | 20 | 4.2 |
| Shale | 3.7 | 12 | 2.7 |
| Average continental crust | 1.42 | 5.6 | 1.43 |
| Chondritic meteorites | 0.008 | 0.029 | 0.056 |

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Parameters (all units same as for the U-238 mentioned below):

## Uranium-238
heat_generation_U238  = 9.46e-5  # W/kg
half_life_U238        = 4.47e9   # years
concentration_U238    = 30.8e-9  # kg/kg
## Uranium-235
heat_generation_U235  = 5.69e-4
half_life_U235        = 7.04e8
concentration_U235    = 0.22e-9
## Thorium-232
heat_generation_Th232 = 2.64e-5
half_life_Th232       = 1.4e10
concentration_Th232   = 124e-9
## Potassium-40
heat_generation_K40   = 2.92e-5
half_life_K40         = 1.25e9
concentration_K40     = 36.9e-9

# Calculate the heat generation through time (4.5 Ga)
# Time range in years
time = np.arange(0, 4.5e9, 100000)

def isotope_heat_generation(t, A0, C, tau):
    return A0 * C * np.exp(t * np.log(2) / tau)

# Calculate heat generation for Uranium-238
A_U238 = isotope_heat_generation(time, heat_generation_U238, concentration_U238, half_life_U238)

# Calculate heat generation for Uranium-235
A_U235 = isotope_heat_generation(time, heat_generation_U235, concentration_U235, half_life_U235)

# Calculate heat generation for Thorium-232
A_Th232 = isotope_heat_generation(time, heat_generation_Th232, concentration_Th232, half_life_Th232)

# Calculate heat generation for Potassium-40
A_K40 = isotope_heat_generation(time, heat_generation_K40, concentration_K40, half_life_K40)

# Plot the heat generation for all isotopes
fig, ax = plt.subplots()
ax.plot(time / 1e9, A_U238, label='238U')
ax.plot(time / 1e9, A_U235, label='235U')
ax.plot(time / 1e9, A_Th232, label='232Th')
ax.plot(time / 1e9, A_K40, label='40K')
ax.plot(time / 1e9, A_U238 + A_U235 + A_Th232 + A_K40, label='Total')
ax.set_xlim(4.5, 0)
ax.set_xlabel('Time (Ga)')
ax.set_ylabel('Heat Generation (W/kg)')
ax.legend()
ax.set_title('Mantle heat Generation vs Time')

plt.show()
```
